```python
def pie_and_bar(y, question = "None"):
  # Tạo một khung ảnh (figure) và hai trục (axes) sắp xếp theo lưới
  fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

  # Biểu đồ tròn cho cột "Cau8"
  y.value_counts().plot.pie(ax=ax1, autopct='%1.1f%%', startangle=90)
  ax1.set_title('Pie Chart for '+ question )

  # Biểu đồ cột thống kê cho cột "Cau8"
  y.value_counts().plot(kind='bar', ax=ax2, color='green', alpha=0.7)
  ax2.set_title('Bar Chart for '+ question)

  # Hiển thị biểu đồ
  plt.show()

pie_and_bar(y, question = "Cau8")
```

# Đúng như ban đầu dữ liệu được chia vào rất nhiều vào hai nhãn 3 và 2  nghĩa là  nhãn khá và tbk

```python
# from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

params = {"n_estimators" : 100,  "max_depth" : 4, \
           "max_leaf_nodes": 3, "learning_rate":0.05, 'min_child_weight': 5}
xgb =  XGBClassifier(**params)
xgb.fit(X_train, y_train)

predict_xgb = xgb.predict(X_test)

print(predict_xgb)
from sklearn.metrics import accuracy_score

accuracy = accuracy_score( y_test, predict_xgb)

# In kết quả
print(f'Accuracy: {accuracy * 100:.2f}%')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Lấy độ quan trọng của các đặc trưng từ mô hình XGBoost đã được điều chỉnh
feature_importance = xgb.feature_importances_

# Lấy tên của các đặc trưng
feature_names = X_train.columns

# Tạo DataFrame để sắp xếp và vẽ biểu đồ
feature_importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': feature_importance})
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

# Vẽ biểu đồ feature importance
plt.figure(figsize=(12, 8))
sns.barplot(x='Importance', y='Feature', data=feature_importance_df, palette='viridis')
plt.title('XGBoost - Feature Importance (Sorted)')
plt.show()
```
