```python
new_columns = ["Gender", "Location", "Birth Year", "Academic Year","University", "Major", " Major Specialization", "Academic Grade",
               "Used ChatGPT", ""]
data_surveys = data_surveys
```

Xử lý câu thứ 15

```python
question_15 = data_surveys["Cau15"].copy()
```

```python
unique_values = question_15.str.split(', ').explode()
# Đếm số lượng các giá trị riêng biệt cho từng quyết định
value_counts = unique_values.value_counts()
print(value_counts)
```

```python
topics = ["Văn hóa",
    "Xã hội", "Điện ảnh", "Thể thao", "Du lịch", "Người nổi tiếng", "Siêu nhiên", "Tâm linh", "Tôn giáo",
    "Sinh hoạt đời thường", "Kiến thức thường nhật", "Kiến thức phổ thông", "Dịch thuật", "Ngôn ngữ"]

one_hot_15 = question_15.str.get_dummies(', ')
result_15 = one_hot_15[topics]
result_15['Khác'] = question_15.apply(lambda x: 1 if all(topic not in x for topic in topics) else 0)
result_15["Ý kiến khác"]  = question_15.apply(lambda x: x if all(topic not in x for topic in topics) else ".")

```

```python
result_15.to_csv("Question_15_onehot.csv", index =True)
```
